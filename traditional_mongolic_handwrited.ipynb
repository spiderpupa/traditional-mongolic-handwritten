```python
f=open('/content/Testset_I_label.txt', 'r', encoding='utf-8')
txt1=f.read()
txt1=txt1.replace('\n', ',')
testLabel_1=txt1.split(',')
print('testLabel_1: ', testLabel_1)

f=open('/content/Testset_II_label.txt', 'r', encoding='utf-16 le')
txt2=f.read()
txt2=txt1.replace('\n', ',')
testLabel_2=txt2.split(',')
print('testLabel_2: ', testLabel_2)

f=open('/content/Trainset_label.txt', 'r', encoding='utf-16 le')
txt3=f.read()
txt3=txt3.replace('\n', ',')
trainLabel=txt3.split(',')
print('trainLabel: ', trainLabel)
```

```python
with open('/content/test1.csv', 'w') as f:
    for i in range(5000):
        f.write(testLabel_1[i]+',')
        for j in range(len(img(i))):
            for k in range(len(img(i)[j])):
                f.write(str(img(i)[j][k])+',')
        f.write('\n')
    f.close()
```
